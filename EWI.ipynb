```python
def ewi(x):
    
    yahoo_financials = YahooFinancials(x)
    data = yahoo_financials.get_historical_price_data(start_date='2007-01-05', 
                                                  end_date='2021-11-26', 
                                                  time_interval='daily')
    df = pd.DataFrame(data[x]['prices'])
    df = df.drop('date', axis=1).set_index('formatted_date')
    return df
```
